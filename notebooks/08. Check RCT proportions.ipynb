```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from mood.rct import get_experimental_configurations
from mood.dataset import MOOD_DATASETS
```

```python
NUMBER_OF_CONFIGS = 100
```

```python
def plot_proportions(dataset):
    cols = ["Algorithm", "Representation", "Train-Val Split", "Criterion", "Seed"]
    configs = get_experimental_configurations(dataset)[:NUMBER_OF_CONFIGS]
    df = pd.DataFrame(configs, columns=cols)
    
    fig, axs = plt.subplots(ncols=len(cols), figsize=(15, 3))
    fig.suptitle(dataset)
    
    for idx, col in enumerate(cols):
        ax = axs[idx]
        df = df.sort_values(col)
        sns.histplot(data=df, x=col, ax=ax)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
        ax.set_ylabel("")
    
    plt.tight_layout()


for dataset in MOOD_DATASETS:
    plot_proportions(dataset)
```
